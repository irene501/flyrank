*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python


print("Decision:")
print("Out of thousands of pages, which ones should a reviewer look at first this week?\n")

print("Who acts, and what they do:")
print("A content or SEO reviewer works through a ranked list and decides what action to take for each page, such as rewriting the content, expanding it, improving the title or metadata, or continuing to monitor it.\n")

print("Output:")
print("A ranked queue of pages, with each page receiving a score, an explanation of why it was flagged, and a suggested action.\n")

print("Cost of a wrong call:")
print("- If the model flags a page that does not need review, the reviewer spends time on the wrong page instead of one that actually needs attention.")
print("- If the model fails to flag a page that is losing traffic despite having real demand, that page may continue losing visitors without being reviewed.")
print("- Since reviewers can only check a limited number of pages, it is most important that the highest-ranked pages are the right ones.\n")

print("Why data or machine learning can help:")
print("No single metric tells the full story. For example, a page may be declining but have very little traffic to recover, or it may rank well in search results but have a poor click-through rate. Machine learning can identify useful patterns by combining several metrics, making it easier to prioritize pages than using a single rule.")
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# 1. How much of the inventory even qualifies as "worth reviewing" under simple rules?
stale_visible = df[(df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)]
declining_with_demand = df[(df["trend_direction"] == "down") & (df["impressions_90d"] >= 100)]
low_ctr_visible = df[(df["impressions_90d"] >= 500) & (df["avg_position"].between(0.01, 20)) & (df["ctr"] < 0.5)]

print(f"stale + visible pages:        {len(stale_visible):>6,}  ({len(stale_visible)/len(df):.2%} of inventory)")
print(f"declining + has demand:       {len(declining_with_demand):>6,}  ({len(declining_with_demand)/len(df):.2%} of inventory)")
print(f"visible + under-capturing CTR: {len(low_ctr_visible):>6,}  ({len(low_ctr_visible)/len(df):.2%} of inventory)")

# 2. Is "declining" evenly spread, or concentrated enough to be worth triaging?
print()
print(df["trend_direction"].value_counts())
print(f"\noverall 'down' rate: {(df['trend_direction']=='down').mean():.2%}")

# 3. The repo's own starter pipeline already ran this lane end-to-end.
# Verified numbers from outputs/model_report.md (client-holdout validation):
print()
print("starter pipeline results (from outputs/model_report.md):")
print(f"{'method':<20}{'ROC AUC':>10}{'precision@50':>15}")
for method, auc, p50 in [
    ("baseline rules", 0.627, 0.240),
    ("logistic regression", 0.700, 0.400),
    ("random forest", 0.750, 0.740),
]:
    print(f"{method:<20}{auc:>10.3f}{p50:>15.3f}")
print(f"\nbaseline gets ~{0.240*50:.0f} of its top 50 right; random forest gets ~{0.740*50:.0f} of its top 50 right.")
```
